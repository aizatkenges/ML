```python
#6
# generate random vector
v = np.random.rand(10)

# norm before
norm_before = np.sqrt(np.sum(v**2))

# normalize
v_normalized = v / norm_before

# norm after
norm_after = np.sqrt(np.sum(v_normalized**2))

print("Vector:", v)
print("\nNorm before:", norm_before)

print("\nNormalized vector:", v_normalized)
print("Norm after:", norm_after)
```

```python
#7
a = np.array([1, 2, 3, 4, 5, 6])
b = np.array([0, -4, 3, -3, 6, 5])

#1.Manual loop
dot_manual = 0
for i in range(len(a)):
    dot_manual += a[i] * b[i]

#2.NumPy
dot_numpy = np.dot(a, b)

print("Manual dot product:", dot_manual)
print("NumPy dot product:", dot_numpy)
```

```python
The dot product is computed as the sum of element-wise multiplications.
Both manual loop and NumPy give the same result: 49.
```

```python
#8
#generate random vectors (length 3)
a = np.random.rand(3)
b = np.random.rand(3)
c = np.random.rand(3)

# left side
left = np.dot(a, b + c)

# right side
right = np.dot(a, b) + np.dot(a, c)

print("a:", a)
print("b:", b)
print("c:", c)

print("\nLeft side a·(b+c):", left)
print("Right side a·b + a·c:", right)

print("\nAre they equal?", np.isclose(left, right))
```

```python
#9 Compute the Euclidean distance matrix between every point in group A and every point in group B.
# Group A (4,2)
A = np.array([
    [2, 1],
    [5, 4],
    [1, 6],
    [7, 2]
])

# Group B (3,2)
B = np.array([
    [0, 0],
    [6, 3],
    [3, 5]
])


A_expanded = A[:, np.newaxis, :]   # (4,1,2)
B_expanded = B[np.newaxis, :, :]   # (1,3,2)

diff = A_expanded - B_expanded

sq = diff ** 2

sum_sq = np.sum(sq, axis=2)

dist = np.sqrt(sum_sq)

print("Distance matrix:\n", dist)
```
